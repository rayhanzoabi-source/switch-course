# Input & Output

```python
## Input, output and strings: 
# 1- ask the user r for their name and print: Hello, !
# 2- Ask for their favorite word and print it backwards (Hint: word[::-1])
# 3- Print the length of that word (Hint: len(word)).
# 4- Ask for a sentence, and print it all in lowercase (Hint: .lower()).

name = input("What is your name?")
print(f" Hello, {name}")

favorite_word = input("What is you favorite name?")
print(f"Your favorite name in backwards is: {favorite_word[::-1]}")

print(f"The length of the word is: {len(favorite_word)} and in lowercase is: {(favorite_word).lower()}")

```

# Conditional exercises

```python
# Conditionals:
# 1- Ask for an integer. Print whether it is: positive / negative / zero

number = input(("Please enter a number"))
number = int(number)

if number == 0:
    print("The number you entered is: Zero")
elif number > 0:
    print("The number you entered is: Positive")    
else:
    print("The number you entered is: Negative")    
```

```python
# Lowercase Check
# 1- Ask the user for a word
# 2- Print “All lowercase” if all letters are lowercase
# 3- Otherwise print “Not all lowercase”

word = input("Type a word please")

if word == word.lower():
    print("All lowercase")
else:
    print("Not all lowercase")    
```

```python
# Vowel check:

letter = input("Please enter a letter:")

letter = letter.lower()

if letter in "aeiou":
    print("Vowel")
else:
    print("Consonant")    
```
