```python
import random

class Student:
    def __init__(self, name, age):
        # Initialize the student's name, age, and empty list of grades
        self.name = name
        self.age = age
        self.grades = []

    def add_grade(self, grade):
        """Add a new grade to the list."""
        self.grades.append(grade)

    def did_fail(self):
        """Return True if any grade is below 60, otherwise False."""
        return any(grade < 60 for grade in self.grades)

# --- Create 2 students ---
student1 = Student("Alice", 20)
student2 = Student("Bob", 22)

# --- Add 5 random grades to each student using random.randint ---
for _ in range(5):
    student1.add_grade(random.randint(0, 100))  # grades from 0 to 100
    student2.add_grade(random.randint(0, 100))

# --- Print student details ---
print(f"Student 1: {student1.name}, Age: {student1.age}, Grades: {student1.grades}")
print(f"Student 2: {student2.name}, Age: {student2.age}, Grades: {student2.grades}")

# --- Check if any student failed ---
if student1.did_fail():
    print(f"{student1.name} has failed ðŸ˜ž")
else:
    print(f"{student1.name} passed! ðŸŽ‰")

if student2.did_fail():
    print(f"{student2.name} has failed ðŸ˜ž")
else:
    print(f"{student2.name} passed! ðŸŽ‰")


```
